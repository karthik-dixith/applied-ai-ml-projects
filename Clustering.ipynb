```python
kmn = KMeans(n_clusters=3,random_state=42)
# create model: clustering
# in this code we suppose 3 clusters

kmn.fit(df[['Alcohol','Malic_Acid','Ash','Ash_Alcanity','Magnesium']])
# fitting data – the input of fitting is a DataFrame with its columns

Clusters = kmn.predict(df[['Alcohol','Malic_Acid','Ash','Ash_Alcanity','Magnesium']])
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram,fcluster
import matplotlib.pyplot as plt
```

```python
df2 = df[['Alcohol','Malic_Acid','Ash','Ash_Alcanity','Magnesium']]
```

```python
hir = linkage(df2, method='complete')
```

```python
plt.figure(figsize=(12,6))
dendrogram(hir, leaf_rotation=90)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Wine Sample')
plt.ylabel('Distance')
plt.show()

# 4. Create flat clusters
labels = fcluster(hir, 4, criterion='distance')  # here 4 is the distance threshold
df['HierCluster'] = labels
print(labels)

# 5. 2D scatter plot (pick any 2 features for visualization)
plt.scatter(df2['Alcohol'], df2['Malic_Acid'], c=labels)
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.title('Wine Clusters from Hierarchical Clustering')
plt.show()
```

```python
from sklearn.cluster import MeanShift

x = df2.values  # convert to numpy array

# 2. Create MeanShift model
ms = MeanShift()
ms.fit(x)

# 3. Get labels and cluster centers
labels = ms.labels_
cluster_centers = ms.cluster_centers_
n_cluster = len(np.unique(labels))
print('Number of estimated clusters:', n_cluster)

# 4. Scatter plot (choose 2 features for visualization)
plt.scatter(df2['Alcohol'], df2['Malic_Acid'], c=labels)
plt.scatter(cluster_centers[:,0], cluster_centers[:,1], marker='x', s=150, linewidth=5, c='red', zorder=10)
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.title('MeanShift Clustering of Wine Dataset')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = df2.values

# 2. Apply DBSCAN
dbscan = DBSCAN(eps=3, min_samples=5)  # you can tune these parameters
dbscan.fit(x)
labels = dbscan.labels_

# 3. Scatter plot (pick 2 features for visualization)
plt.scatter(df2['Alcohol'], df2['Malic_Acid'], c=labels)
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.title('DBSCAN Clustering of Wine Clustering')
plt.show()

# 4. Heatmap of correlations
plt.figure(figsize=(8,6))
sns.heatmap(df2.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap of Wine Clustering')
plt.show()
```
